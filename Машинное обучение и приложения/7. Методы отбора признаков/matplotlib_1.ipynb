```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1,2,3], [1,4,9])   # pyplot помнит текущую фигуру
plt.title("Квадраты")         # заголовок добавится к ней же
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = [1,2,3]
plt.plot(x, [1,4,9], 'r-', label="y = x²")
plt.plot(x, [1,2,3], 'b--', label="y = x")
plt.legend()   # теперь в углу появится табличка с объяснением
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
plt.scatter(x, y, color='red', s=100)
plt.show()
```

```python
categories = ['А', 'Б', 'В']
values = [15, 30, 45]
plt.bar(categories, values, color=['blue', 'green', 'orange'])
plt.show()
```

```python
data = np.random.randn(1000)  # нормальное распределение
plt.hist(data, bins=20, edgecolor='black')
plt.show()
```
