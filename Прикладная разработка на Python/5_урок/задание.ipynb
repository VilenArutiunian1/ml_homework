```python
# 18

data_copy = data.copy()

data_copy["dateUpdated"] = pd.to_datetime(data_copy["dateUpdated"])
data_copy["dateAdded"] = pd.to_datetime(data_copy["dateAdded"])

data_copy["update_delta"] = (data_copy["dateUpdated"] - data_copy["dateAdded"]).dt.days
print(data_copy["update_delta"].mean(), data_copy["update_delta"].median())
```

```python
# 19


data_copy = data.copy()
data_copy["categoriesCount"] = data["categories"].str.split(',').apply(lambda x: len(x))

result = data_copy[data_copy["categoriesCount"] > 20].groupby("province").agg({"longitude": "min"}).round(3)
result.to_csv("csv_file.csv", sep=';')

```
